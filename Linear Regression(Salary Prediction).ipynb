```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Salary Data.csv")
df
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary Data')
```

```python
x = df.drop('Salary',axis = 1)
```

```python
x.head()
```

```python
y = df.drop('Experience Years',axis = 1)
```

```python
y.head()
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color = "red")
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary Data')
```

```python
reg.fit(x,y) #train ML model
```

```python
m = reg.coef_
```

```python
c = reg.intercept_
```

```python
# y = mx + c
result = m*10.7 + c
```

```python
result
```

```python
reg.predict([[10.7]])
```

```python
df['Predicted_y'] = reg.predict(x)
df
```

```python
plt.plot(x,df['Predicted_y'])
plt.scatter(x.mean(),y.mean(),color = "red")
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary Data')
```

```python
reg.predict([[10.9]])
```

```python
df.head()
```

```python
df
```

```python
# Lost and Cost Calculation
df['loss'] = df['Salary'] - df['Predicted_y']
```

```python
df.head()
```

```python
# MSE(Mean Squre Error) and MAE(Mean Absolute Error)
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
MSE = mean_squared_error(df['Salary'],df['Predicted_y'])
MSE
```

```python
MAE = mean_absolute_error(df['Salary'],df['Predicted_y'])
MAE
```

```python
#another way to calculation mse
sum(abs(df['loss']))/len(x)
```

```python
#Now calculated acquracy
reg.score(x,y)
```

```python
from sklearn.metrics import r2_score
r2_score(reg.predict(x),y)
```
